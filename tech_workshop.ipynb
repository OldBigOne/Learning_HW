```python
#Завдання 4. Написати функцію, яка приймає рядок та повертає кількість унікальних символів у цьому рядку:
def symbol_count (x):
  count=[]
  for i in x:
    if i not in count:
      count.append(i)
  print ('Кількість унікальних символів: ',len(count))

x = input('Введіть рядок:')
symbol_count (x)
```

```python
#PANDAS Завдання 5. Використати функцію apply для створення нового стовпця на основі умови - У вас є таблиця з оцінками студентів,
# і вам потрібно створити новий стовпець "Grade" на основі наступних критеріїв: A (90 і вище), B (80-89), C (менше 80).

import pandas as pd
data = {'Score': [85, 92, 78, 95, 88]}
df = pd.DataFrame(data)
df.index = df.index = [f'Student {i+1}' for i in range(len(df))]

def rating(score):
  if score <80:
    return 'C'
  elif score <89:
    return 'B'
  else:
    return 'A'

df['Grade'] = df['Score'].apply(rating)
print(df)
```
